## Simple example of using a <b>while</b> loop
The semicolon directs everything in the conditional, and proceedding indented block, to execute while the condition is true

```python
i = 0
while i < 5:
    print(i)
    i+=1
```

<br>
 
 ## Simple examples of using a <b>for</b> loop
 The semicolon directs everything in the conditional, and proceedding indented block, to execute while the condition is true

```python
for i in range(3):
    print(i)
```

```python
for c in "python":
    print(c)
```

```python
days=["Mon","Tues","Wed","Thur","Fri","Sat","Sun"]
for d in days:
    if d != "Wed":
        print(d)
```

<br>

## Simple example of using <b>if elif else</b> conditionals. The semicolon operates the same as <b>while</b> and <b>for</b>

```python
i = 10
if i < 20:
    print(i,"is less than 20")
elif i > 20:
    print(i,"is greater than 20")
else:
    print("i is 20")
```
